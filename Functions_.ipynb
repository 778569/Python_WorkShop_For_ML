**Python Functions can return multiple values**

Python Functions has return multiple returns, those are returned to tuple

---



```python
def Calculator(val1, val2):
  add= val1+ val2
  sub= val1- val2
  mult= val1* val2
  div= val1/ val2
  return add,sub,mult,div
```

```python
add,sub,mult,div = Calculator(10,20)
print(add)
print(sub)
print(mult)
print(div)
```
